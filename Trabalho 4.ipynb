```python
from pybrain.tools.shortcuts import buildNetwork
from pybrain.supervised.trainers import BackpropTrainer
from pybrain.datasets import SupervisedDataSet
import numpy as np
from math import ceil
from matplotlib import pyplot as pl
%matplotlib inline
```

```python
def boundary(net, X, y, ax, inc=1., alpha=.15, step=.1):
    
    x1_min, x1_max = X[:, 0].min()-inc, X[:, 0].max()+inc
    x2_min, x2_max = X[:, 1].min()-inc, X[:, 1].max()+inc
    
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, step),
                         np.arange(x2_min, x2_max, step))

    xx = np.c_[xx1.ravel(), xx2.ravel()]
    
    ds = SupervisedDataSet(X.shape[1], y.shape[1])
    ds.setField('input', xx)
    ds.setField('target', np.zeros_like(xx))
    
    Z = net.activateOnDataset(ds)
    
    Z = Z.reshape(xx1.shape)
    
    Z = np.where(Z < 0, -1., 1.)

    ax.contourf(xx1, xx2, Z, alpha=alpha, c=('r', 'b'))
    ax.scatter(X[:, 0], X[:, 1], c=y)
```

```python
def contour(filepath, hiddens, max_epochs=10):
    
    n_plots = len(hiddens)
    n_rows = int(ceil(n_plots/3.))
    n_columns = 3

    # plot
    fig, axs = pl.subplots(n_rows, n_columns, figsize=(9, n_rows*3))

    # dados
    dados = np.genfromtxt(filepath)
    X, y = dados[:, :-1], dados[:, -1].reshape((dados.shape[0], -1))

    ds = SupervisedDataSet(X.shape[1], y.shape[1])
    ds.setField('input', X)
    ds.setField('target', y)

    for hidden, ax in zip(hiddens, axs.ravel()):
        
        estrutura = [X.shape[1]]
        estrutura.extend(hidden)
        estrutura.append(y.shape[1])

        net = buildNetwork(*estrutura)
        trainer = BackpropTrainer(net, ds)

        trainer.trainUntilConvergence(maxEpochs=max_epochs)

        boundary(net, X, y, ax=ax)    
        ax.set_title('hidden %s' % ','.join([str(h) for h in hidden]))
```
