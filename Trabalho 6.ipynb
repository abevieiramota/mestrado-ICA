```python
import numpy as np
from matplotlib import pyplot as pl
import random
import time
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
def ackley(x, y):
    
    return -20.*np.exp(-.2*np.sqrt(.5*(x**2 + y**2))) - np.exp(.5*(np.cos(2*np.pi*x) + np.cos(2*np.pi*y))) + 20. + np.e
```

```python
fig = pl.figure()
ax = fig.gca(projection='3d')
X = np.linspace(-30, 30, 300)
Y = np.linspace(-30, 30, 300)
X, Y = np.meshgrid(X, Y)

Z = ackley(X, Y)

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

pl.show()
```

```python
def selecionar_aptos(X, Y, qtd):
    
    Z = ackley(X, Y)
    
    Z = Z.reshape((Z.shape[0], 1))
    
    Z = np.hstack((np.arange(0, len(Z)).reshape((len(Z), 1)), Z))
    
    Z = Z[Z[:, -1].argsort()]
    
    i_ordered = Z[:qtd, 0].astype(np.int)
    
    return np.hstack((X[i_ordered], X[i_ordered])), np.hstack((Y[i_ordered], Y[i_ordered]))
    

def cruzar(X):
    
    random.seed(time.time())
    mid = len(X)/2
    resultado = []
    # crossover
    for x, y in zip(X[:mid], X[mid:]):

        # representação binária
        
        x_2 = [int(w) for w in bin(x).split('b')[1]]
        y_2 = [int(w) for w in bin(y).split('b')[1]]
        
        # ponto de corte
        len_min = max(len(x_2), len(y_2))
        base = '%.'+str(len_min)+'d'
        x_2 = [int(w) for w in base % int(''.join([str(w) for w in x_2]))]
        y_2 = [int(w) for w in base % int(''.join([str(w) for w in y_2]))]
        ponto_de_corte = random.choice(range(len_min))
        
        # troca
        x_2_l = x_2[ponto_de_corte:]
        x_2_r = x_2[:ponto_de_corte]
        y_2_l = y_2[ponto_de_corte:]
        y_2_r = y_2[:ponto_de_corte]
        
        x_2_l.extend(y_2_r)
        y_2_l.extend(x_2_r)
        resultado.append(x_2_l)
        resultado.append(y_2_l)
        
    # mutação
    for x in resultado:
        
        if random.randint(0, 100) == 0:
        
            pos = random.randint(0, len(x)-1)
            if x[pos] == 0:
                x[pos] = 1
            else:
                x[pos] = 0
            
    # to int
    return np.array([int(''.join([str(w) for w in x]), 2) for x in resultado])
```

```python
%%time
SIZE = 8
MAX_EPOCHS = 5000

X = np.random.randint(1000, 9000, SIZE)
Y = np.random.randint(1000, 9000, SIZE)

Xs = []
Ys = []

i = 0
aptidao_min = 1.
A_min = []
A_mean = []
A_max = []
A_median = []
while aptidao_min > 0 and i < MAX_EPOCHS:
    
    i += 1
    aptos_X, aptos_Y = selecionar_aptos(X, Y, SIZE/2)
    
    filhos_X, filhos_Y = cruzar(aptos_X), cruzar(aptos_Y)
    
    X, Y = filhos_X, filhos_Y
    
    Z = ackley(X, Y)
    
    aptidao_min = np.min(Z)
    A_min.append(aptidao_min)
    A_mean.append(np.mean(Z))
    A_max.append(np.max(Z))
    A_median.append(np.median(Z))
    
    Xs.append(X)
    Ys.append(Y)
```

```python
fig = pl.figure(figsize=(16, 4))
epocas = range(i)

ax1 = fig.add_subplot(1, 4, 1)
ax1.plot(epocas, A_min)
ax1.set_title('Min')

ax2 = fig.add_subplot(1, 4, 2)
ax2.plot(epocas, A_mean)
ax2.set_title('Mean')

ax3 = fig.add_subplot(1, 4, 3)
ax3.plot(epocas, A_max)
ax3.set_title('Max')

ax4 = fig.add_subplot(1, 4, 4)
ax4.plot(epocas, A_median)
ax4.set_title('Median')
```
