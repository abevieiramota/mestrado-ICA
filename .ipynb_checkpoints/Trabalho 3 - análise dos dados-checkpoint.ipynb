```python
import os
import matplotlib.pyplot as pl
import numpy as np
from math import ceil
import pandas as pd
%matplotlib inline
```

```python
# filepath
DADOS_FOLDER = 'Dados'
DADOS_FILENAME = 'abalone.data'
DADOS_FILEPATH = os.path.join(DADOS_FOLDER, DADOS_FILENAME)

# colunas
DADOS_COLUMNS = ['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight', 'Rings']
# mapeamento de rótulos para números
SEX_TO_INT = {"M":-1., "F":1., "I":0.}
```

```python
dados = pd.read_csv(DADOS_FILEPATH, 
                    names=DADOS_COLUMNS, 
                    converters={0:lambda x: SEX_TO_INT[x]})

features = dados.columns[:-1]
pares = [(features[i], features[j]) for i in xrange(len(features)-1) for j in xrange(i+1, len(features))]
```

### Exemplos dos dados

```python
dados[:10]
```

### Estatística dos dados

```python
dados.describe()
```

### Correlação

```python
corr_vals = [dados.corr()[f1][f2] for (f1, f2) in pares]
corrs = pd.DataFrame(corr_vals, pares, columns=['correlação'])
```

### Boxplot

```python
fig = pl.figure(figsize=(8, 4))
ax = fig.add_subplot(1,1,1)

_ = corrs.sort('correlação').plot(kind='bar', ax=ax, rot=90)
```

### Distribuição dos dados

```python
n = len(dados.columns)
fig, axs = pl.subplots(nrows=int(ceil(n/3.)), ncols=3, figsize=(9, n))

for ax, col in zip(axs.ravel(), dados.columns):
    
    dados[col].hist(ax=ax)
    ax.set_title(col)
```

### Scatter dos dados, utilizando duas features por vez

```python
n = len(pares)
fig, axs = pl.subplots(nrows=int(ceil(n/3.)), ncols=3, figsize=(9, n))

for (col0, col1), ax in zip(pares, axs.ravel()):
        
    ax.scatter(dados[col0], dados[col1], dados['Rings'])
    ax.set_title('%s - %s' % (col0, col1))
```

```python
fig = pl.figure(figsize=(10, 10))
ax = fig.add_subplot(1,1,1)

_ = dados.boxplot(ax=ax, rot=30)
```
