```python
from pybrain.tools.shortcuts import buildNetwork
from pybrain.supervised.trainers import BackpropTrainer
from pybrain.datasets import SupervisedDataSet
import numpy as np
from matplotlib import pyplot as pl
%matplotlib inline
```

```python
!ls Dados/T4/
```

```python
dados = np.genfromtxt("Dados/T4/circular.dat")
```

```python
X, y = dados[:, :-1], dados[:, -1].reshape((y.shape[0], -1))
ds = SupervisedDataSet(X.shape[1], y.shape[1])
ds.setField('input', X)
ds.setField('target', y)

net = buildNetwork(X.shape[1], 5, y.shape[1])
trainer = BackpropTrainer(net, ds)

trainer.trainUntilConvergence(maxEpochs=10)
```

```python
def boundary(net, X, Y, inc=1., alpha=.15, step=.1):
    
    x1_min, x1_max = X[:, 0].min()-inc, X[:, 0].max()+inc
    x2_min, x2_max = X[:, 1].min()-inc, X[:, 1].max()+inc
    
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, step),
                         np.arange(x2_min, x2_max, step))

    xx = np.c_[xx1.ravel(), xx2.ravel()]
    
    ds = SupervisedDataSet(X.shape[1], y.shape[1])
    ds.setField('input', xx)
    ds.setField('target', np.zeros_like(xx))
    
    Z = net.activateOnDataset(ds)
    
    Z = Z.reshape(xx1.shape)
    
    Z = np.where(Z < 0, -1., 1.)

    pl.contourf(xx1, xx2, Z, alpha=alpha, c=('r', 'b'))
    pl.scatter(X[:, 0], X[:, 1], c=Y)
    
    return Z
```

```python
Z = boundary(net, X, y)
```
