```python
import numpy as np
from matplotlib import pyplot as pl
%matplotlib inline
```

```python
def gaussian(a, b, c, d):
    
    def f(x):
        
        return a*np.exp(-1.*((x-b)**2)/(2*c**2)) + d

    return f

rules = {
"LE":{
    "RB":"PS",
    "RU":"NS",
    "RV":"NM",
    "VE":"NM",
    "LV":"NB",
    "LU":"NB",
    "LB":"NB"
},
"LC":{
    "RB":"PM",
    "RU":"PS",
    "RV":"NS",
    "VE":"NM",
    "LV":"NM",
    "LU":"NB",
    "LB":"NB"
},
"CE":{
    "RB":"PM",
    "RU":"PM",
    "RV":"PS",
    "VE":"ZE",
    "LV":"NS",
    "LU":"NM",
    "LB":"NM"
},
"RC":{
    "RB":"PB",
    "RU":"PB",
    "RV":"PM",
    "VE":"PM",
    "LV":"PS",
    "LU":"NS",
    "LB":"NM"
},
"RI":{
    "RB":"PB",
    "RU":"PB",
    "RV":"PB",
    "VE":"PM",
    "LV":"PM",
    "LU":"PS",
    "LB":"NS"
}
}

def limited_fun(fun, x, limit):
    
    y = fun(x)
    y[y>limit] = limit
    
    return y    
```

# Item I

```python
# Distância
distancia = {
    "CE": gaussian(1., 50., 2., 0.),
    "LC": gaussian(1., 40., 3., 0.),
    "RC": gaussian(1., 60., 3., 0.),
    "LE": gaussian(1., 0., 10., 0.),
    "RI": gaussian(1., 100., 10., 0.)
}

x = np.linspace(0, 101, 200)
for d, g in distancia.items():
    
    pl.plot(x, g(x), label=d)
    
pl.legend()
pl.xlim((-1, 101))
pl.ylim((0, 1.2))
pl.title(u"distância")
```

```python
# Ângulo de entrada

angulo_in = {
    "VE": gaussian(1., 90., 5., 0.),
    "RV": gaussian(1., 65., 15., 0.),
    "LV": gaussian(1., 115., 15., 0.),
    "RU": gaussian(1., 30., 5., 0.),
    "LU": gaussian(1., 165., 5., 0.),
    "RB": gaussian(1., -20., 50., 0.),
    "LB": gaussian(1., 225., 50., 0.)
}

x = np.linspace(-100, 300, 500)
for d, g in angulo_in.items():
    
    pl.plot(x, g(x), label=d)
    
pl.legend()
pl.xlim((-100, 300))
pl.ylim((0, 1.2))
pl.title(u"ângulo de entrada")
```

```python
# Ângulo de saída

angulo_out = {
    "ZE": gaussian(1., 0., 2., 0.),
    "NS": gaussian(1., -5., 2., 0.),
    "PS": gaussian(1., 5., 2., 0.),
    "NM": gaussian(1., -15., 4., 0.),
    "PM": gaussian(1., 15., 4., 0.),
    "NB": gaussian(1., -30., 7., 0.),
    "PB": gaussian(1., 30., 7., 0.)
}

x = np.linspace(-30, 30, 200)
for d, g in angulo_out.items():
    
    pl.plot(x, g(x), label=d)
    
pl.legend()
pl.xlim((-30, 30))
pl.ylim((0, 1.2))
pl.title(u"ângulo de saída")
```

# Item II

- distancia = 47.5 
- ângulo = 99

```python
x = 47.5
a_in = 99

x_perts = [(d_name, d(x)) for (d_name, d) in distancia.items()]
a_in_perts = [(a_in_name, a_in_fun(a_in)) for (a_in_name, a_in_fun) in angulo_in.items()]

matriz = {d_name:{a_in_name: min(a_in_value, d) for (a_in_name, a_in_value) in a_in_perts} for (d_name, d_value) in x_perts}

a_out_limits = {k:0. for k in angulo_out.keys()}

for d in matriz.keys():
    
    for a in matriz[d].keys():
        
        a_out = rules[d][a]
        a_out_limits[a_out] = max(a_out_limits[a_out], matriz[d][a])
        
# Ângulo de saída

x = np.linspace(-30, 30, 100)
for d, g in angulo_out.items():
    
    pl.plot(x, limited_fun(g, x, a_out_limits[d]), label=d)
    
pl.legend()
pl.xlim((-30, 30))
pl.ylim((0, 1.2))
pl.title(u"aplicação da regra")
```
